```python
from google_play_scraper import *
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from operator import itemgetter
import nltk
nltk.download('punkt')
N = 5
pd.set_option('display.max_colwidth', None)
```

```python
result = search(
    "Best Music Apps",
    lang="en",  # defaults to 'en'
    country="us",  # defaults to 'us'
    n_hits=100  # defaults to 30 (= Google's maximum)
)
```

```python
all_apps = []
app_ids = {}
for each in result:
    all_apps.append((each['title'],int(each['installs'][:-1].replace(",",""))))
    app_ids[each['title']] = each['appId']
    print('App Name: {} Downloads:{}'.format(each['title'],each['installs'][:-1]))
top_n_apps = sorted(all_apps, key = itemgetter(1), reverse = True)[:N]
```

```python
top_n_apps
```

```python
for each in top_n_apps:
    print("App: ",each[0])
    print("App ID: ",app_ids[each[0]])
    if app_ids[each[0]] != 'com.strava' or app_ids[each[0]] != 'com.sec.android.app.shealth' or app_ids[each[0]] != 'com.google.android.apps.fitness':
        user_reviews = reviews_all(
        app_ids[each[0]],
        sleep_milliseconds=0,
        lang='en', # Default language is 'en'
        country='us', # Default country is 'us'
        sort=Sort.NEWEST, # Default is Sort.MOST_RELEVANT.
        )
        df_reviews = pd.DataFrame(np.array(user_reviews),columns=['review'])
        df_reviews = df_reviews.join(pd.DataFrame(df_reviews.pop('review').tolist()))
        df_reviews.to_csv('music/'+(each[0].replace(' ',''))+'.csv',header=True)
```

```python
filtered_dict_list = []
```

```python
app_review_count = 0
for review in dict_list:
    # Preprocess the review text by converting it to lowercase and tokenizing it into individual words
    review_text = str(review['content']).lower()
    review_words = word_tokenize(review_text)
    
    # Check if any of the keywords are present in the review
    if any(keyword in review_words for keyword in accessibility_keywords+browser_keywords):
        app_review_count += 1
        filtered_dict_list.append(review)
```
